*  __Abstract:__
<br>
Over 91,000 adverse food events were reported to the FDA between 2004 and the first quarter of 2021, with the largest amounts of daily reporting likely resulting from publicly announced law suits, and public health issues. Twice as many women as men were filing reports when gender was declared, with those most likely to report being over that age of 50. The Vitamin/Miniral/Protein/Unconventional/Diet industry and the Cosmetic industry were given the most reports of adverse reactions and outcomes by far in percentiles above average being over 2000%. The most reported reactions included diarrhoea, vomiting, nausea, and abdominal pain, but at the 3rd most reported, Ovarian Cancer was looked into more deeply into and appeared to be tied to Johnson & Johnson's Baby Powder. The most likely outcome was medical care, while "Other Outcome" was a second, likely due to it being a generic catch-all for those who did not fit the other 10 outcomes listed. Both men and women shared the top three adverse reactions of diarrhoea, nausea, and vomiting, but while men went on to have abdominal pain and melaise, women had the more serious reactions of ovarian cancer and choking rounding out their top five. Both men and women shared the top 5 of the 11 outcomes, with men being more likely to be hospitalized before seeing a doctor. Death as an outcome differed vastly between the the two, however. Women were twice as likely to have death as an outcome as compared to men. Further investigation is needed, however, to determine the connections between death, gender, industry, and reactions, if there are any.

<br><br>
*  __Information About the Source Data__
<br>
Data was sourced via the Open FDA API for adverse food event reports (https://open.fda.gov/apis/food/event/). The data is made available to the FDA by The Center for Food Safety and Applied Nutrition Adverse Event Reporting Database, which contains reports of adverse events from consumers, health professionals, and members of the food industry. Likely due to HIPPA regulations, the reports contain only the report number, date reported, date the event began, reactions, outcome, industry, and product; age, gender, and product name are not always supplied. Available data is dated from 1 January 2004 to 30 March 2021, but does not undergo extensive validation or verification; relationships between reactions/outcomes to particular products can be suspected, but not required. 
<br><br>
*   __Adverse Reports by Day Summary__
<br>
Between 1 January 2004 and 30 March 2021 a __total of 91,776 reports were filed__. The minimum filed in one day was 1, while the maximum was 204 on 18 April 2017. An average day saw 12 reports being filed, but skewed towards higher numbers as time progressed. The 20 days with the most reporting were primarily the result of four things. 18 April 2017, 171 of the 204 reports cited Quorn Food Products, and was likely a direct reaction to a class action lawsuit against Quorn Food Products for misleading advertising; the protein replacement food had not been made by mushrooms as they had been lead to believe, but instead firmented mold. On 21 February 2020, 260 vitamin products were reported on as a likely outcome of the public's first reaction to the arrival of COVID in the United States. On 13 March 2018, when 165 if the 194 reports filed that day with Ovarian Cancer as a reaction to baby powder, it was as a result of law suits against Johnson & Johnson which revealed that their talcum powder contained asbestos. 4 May 2011 saw 103 of 115 reports made citing Hydroxycut, a weightloss suppliment, but cursory research of news media turned up nothing as to why so many reports had been made that day in regards to the product.

## 4. Count of Reports Filed By Day: Statistics and Graph

```python
### PULL & STATISTICS: Count of Reports Filed Per Day

# Pull report from the API
req = requests.get('https://api.fda.gov/food/event.json?count=date_created') # <--request data by date_created from API
rept_by_date_data = req.json()                   # <-- put requested data into JSON (JavaScript Object Notation) format

# Look at incoming data for the 'results' section to pull the reaction and count numbers and put in the dictionary   
if(('results' in rept_by_date_data) and (bool(rept_by_date_data['results']) == True)):
    for inpt in rept_by_date_data['results']:              # <-- the header for the API is 'results'
        report_by_date_dict['date'].append(inpt['time'])   # <-- read 'time' from API data and put into 'date' of dictinary
        report_by_date_dict['count'].append(inpt['count']) # <-- read 'count' from API data and put into 'count' of dictionary 

# Create Data Frame from Dictionary
df_report_by_date = pd.DataFrame.from_dict(report_by_date_dict)

# Describe the Data (Statistics): and round to the 10ths
np.round(df_report_by_date.describe(), 1) 
```

```python
# Total Reports Submitted
df_report_by_date['count'].sum()
```

```python
### GRAPH: Number of Reports Filed Per day

fig, ax = plt.subplots(1, 1)

# Convert the 'date' column in the dictionary from text to date format
df_report_by_date['date'] = pd.to_datetime(df_report_by_date['date'], format= '%Y%m%d')

# The plot line graph
df_report_by_date.plot.line (ax = ax, x = 'date', figsize = (16,8), fontsize=16, color = 'tomato', legend= False)
plt.gca().grid(True, linewidth=0.7, linestyle=':')
plt.xlabel('Dates', fontsize = 20)
plt.ylabel('Number of Reports', fontsize = 20)
plt.title('How many reports came in per day?: 2004 - 2021/03/31', fontsize = 30)

# Saving an image of the graph
#plt.savefig('daily_report_counts.png',bbox_inches = 'tight')
```
